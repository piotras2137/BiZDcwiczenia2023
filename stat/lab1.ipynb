# zadanie 2

```python
import statistics as st

with open('csv/Wzrost.csv', 'r') as f:
    data = [float(x) for x in (f.read().splitlines())[0].split(',')]

print('mean: ', st.mean(data))
print('fmean', st.fmean(data))
print('geometric mean', st.geometric_mean(data))
print('harmonic mean', st.harmonic_mean(data))
print('median: ', st.median(data))
print('median low: ', st.median_low(data))
print('median high: ', st.median_high(data))
print('median group: ', st.median_grouped(data))
print('mode: ', st.mode(data))
print('pvariance: ', st.pvariance(data))
print('variance: ', st.variance(data))
print('pstdev: ', st.pstdev(data))
print('stdev: ', st.stdev(data))
print('quantiles: ', st.quantiles(data))
print('standard error: ', st.stdev(data) / len(data)**0.5)
print('standard error of the mean: ', st.stdev(data) / len(data)**0.5)
print('standard deviation of the mean: ', st.stdev(data) / len(data)**0.5)
```

# zadanie 3

```python
df = pd.read_csv('csv/Waga.csv', sep=';', na_values='.')
df
```

```python
from scipy import stats
import numpy as np

# Statystyki opisowe
descriptive_stats = stats.describe(df['Åšrednia Waga (kg)'])
print("Statystyki opisowe:")
print(descriptive_stats)

# Dominanta
mode_result = stats.mode(df['Åšrednia Waga (kg)'])
print("\nDominanta:")
print(mode_result)

# SkosnoÅ›Ä‡ i Kurtoza
skewness = stats.skew(df['Åšrednia Waga (kg)'])
kurtosis = stats.kurtosis(df['Åšrednia Waga (kg)'])
print("\nSkosnoÅ›Ä‡:", skewness)
print("Kurtoza:", kurtosis)

# Test normalnoÅ›ci (test Shapiro-Wilka)
shapiro_test_stat, shapiro_test_p = stats.shapiro(df['Åšrednia Waga (kg)'])
print("\nTest Shapiro-Wilka:")
print(f"Statystyka testowa: {shapiro_test_stat}, WartoÅ›Ä‡ p: {shapiro_test_p}")

# PrzedziaÅ‚ ufnoÅ›ci dla Å›redniej
confidence_interval = stats.t.interval(0.95, len(df['Åšrednia Waga (kg)'])-1, loc=np.mean(df['Åšrednia Waga (kg)']), scale=stats.sem(df['Åšrednia Waga (kg)']))
print("\n95% PrzedziaÅ‚ ufnoÅ›ci dla Å›redniej:")
print(confidence_interval)

# Test hipotezy dla Å›redniej (test t dla jednej prÃ³by)
t_stat, t_p_value = stats.ttest_1samp(df['Åšrednia Waga (kg)'], popmean=5.0)
print("\nTest t dla jednej prÃ³by:")
print(f"Statystyka t: {t_stat}, WartoÅ›Ä‡ p: {t_p_value}")
```

```python
df = pd.read_csv('csv/brain_size.csv', sep=';', na_values='.')
df = df.drop(df.columns[0], axis=1)

# UzupeÅ‚nianie pustych danych
mean_weight = df['Weight'].mean()
mean_height = df['Height'].mean()
df['Weight'].fillna(mean_weight, inplace=True)
df['Height'].fillna(mean_height, inplace=True)

df
```

```python
print('Mean (VIQ):', df['VIQ'].mean())
print('Male count:', len(df[df['Gender']=='Male']))
print('Female count:', len(df[df['Gender']=='Female']))
```

```python
import matplotlib.pyplot as plt

columns = ['VIQ', 'PIQ', 'FSIQ']
for column in columns:
    plt.hist(df[column], bins=20)
    plt.title(column)
    plt.show()
```

```python
df_female = df[df['Gender']=='Female']
for column in columns:
    plt.hist(df_female[column], bins=20)
    plt.title(f'{column} (female)')
    plt.show()
```
